Procesar el archivo y extraer:

Cantidad de vocales

Cantidad de consonantes

Listado de las 50 palabras que más se repiten

```python
## Se crea una funcion principal que analiza el codigoi
def realizar_analisis(archivo):
    #Con estas dos variables se compara el archivo, para determinar vocales y consonantes 
    vocales = "aeiouáéíóú"
    consonantes = "bcdfghjklmnñpqrstvwxyz"
    #Esto apre el archivo y lo decodifica con ayuda de utf-8
    with open(archivo, "r", encoding="utf-8") as archiv:## 
        texto_analizar = archiv.read().lower()## Aqui se pone todo en minucuclas para facilitar el analisis 
    
    #Se crean 2 variables que van a lleavr la cuenta 
    contador_vocales = 0
    contador_consonantes= 0
    ## Este for recorre el texto y compara si una vocal es coincidente con el elemento en el que este posicionado, si es asi hace la sumatoria 
    for palabra in texto_analizar:
        if palabra in vocales:
            contador_vocales+=1 
     ## Este for recorre el texto y compara si una consonante es coincidente con el elemento en el que este posicionado, si es asi hace la sumatoria 
    for palabra2 in texto_analizar:
        if palabra2 in consonantes:
            contador_consonantes+=1 

    # Se imprimen los resultados
    print("La cantidad de vocales es:", contador_vocales)
    print("La cantidad de consonantes es:", contador_consonantes)

    # Se didvide el texto para analizar palabra por palabra 
    palabras_separadas = texto_analizar.split()
    
    # Crear un diccionario para contar la frecuencia de cada palabra
    frecuencia_palabras = {}
    for palabra in palabras_separadas:
        if palabra in frecuencia_palabras: #Si la palabra ya existe en el diccionario, significa que ya apareció antes.
            frecuencia_palabras[palabra] += 1
        else:
            frecuencia_palabras[palabra] = 1 # Si la iteracion pasa y no se ha repetido le da un valor inicial de 1 
    
    # Convertir el diccionario en una lista de tuplas (palabra, frecuencia)
    lista_palabras_frecuencia = list(frecuencia_palabras.items())
    
    # Función para ordenar por frecuencia 
    def obtener_frecuencia(tupla):
        return tupla[1]  # Retorna el segundo elemento de la tupla (la frecuencia)
    # Ordenar la lista de tuplas por frecuencia (de mayor a menor)
    lista_palabras_frecuencia.sort(key=obtener_frecuencia, reverse=True)
    # Seleccionar las 50 palabras más comunes
    palabras_mas_comunes = lista_palabras_frecuencia[:50]
    
    # Imprimir las 50 palabras más comunes
    print("Las 50 palabras más comunes son:")
    for palabra, frecuencia in palabras_mas_comunes:
        print(palabra,":", frecuencia)
## Esta es la ubicacion del archivo
ruta = r"C:\Users\migue\Downloads\archivo_de_reto_11.txt"
#Se usa la funcion 
realizar_analisis(ruta)
```
